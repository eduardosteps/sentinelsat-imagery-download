```python
from sentinelsat import SentinelAPI, read_geojson, geojson_to_wkt
from datetime import date
```

```python
# connect to the API
user = '<your_username>'
passwd = '<your_password>'
api = SentinelAPI(user, passwd, 'https://scihub.copernicus.eu/dhus')
```

```python
# search by polygon, time, and Hub query keywords
footprint = geojson_to_wkt(read_geojson("./geojson/salvador-ba.geojson"))
products = api.query(footprint,
                     date = ('20200101', '20200331'),
                     platformname = 'Sentinel-2',
                     processinglevel = 'Level-2A',
                     cloudcoverpercentage = (0, 10))
#print(footprint)
#print(products)
len(products)
```

```python
# For loop to download the objects one by one
for i in products:
    api.get_product_odata(i)
    api.download(i)
```
